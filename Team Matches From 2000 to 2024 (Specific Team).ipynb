```python
from bs4 import BeautifulSoup
import time
import requests
import pandas as pd
import numpy as np
import re
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36'
}
Team_ID = []
Team_Name = []
base_url = "https://www.transfermarkt.com/egyptian-premier-league/startseite/wettbewerb/EGY1"
pageTree = requests.get(base_url, headers=headers)
soup = BeautifulSoup(pageTree.content, 'html.parser')
teams = soup.find_all('td',{'class':'hauptlink no-border-links'})
for team in teams:
  if team.find('a'):
    href = team.find('a')['href']
    team_id = href.split('/verein/')[1].split('/')[0]
    Team_ID.append(team_id)
    Team_Name.append(team.text.strip())
Teams_Data = dict(zip(Team_ID, Team_Name))
```

```python
# Initialize main lists to store data
Score = []
Coach = []
Match_Round = []
Match_Date = []
Match_Time = []
Venue = []
Formation = []
Away = []
Home = []
Season = []
season = 2000
headers = {
    'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36'
}

while True:
    # Temporary lists for each season
    temp_round = []
    temp_date = []
    temp_time = []
    temp_date = []
    temp_time = []
    temp_away = []
    temp_formation = []
    temp_home = []

    #Team Id "Ahly = 7"  && "Zamalek = 667"
    base_url = f"https://www.transfermarkt.co.uk/el-ahly-kairo/spielplan/verein/7/saison_id/{season}/plus/1#EGY1"
    pageTree = requests.get(base_url, headers=headers)
    soup = BeautifulSoup(pageTree.content, 'html.parser')

    rows = soup.find_all('tr')

    # Extract Score and Coach
    for row in rows:
        cells = row.find_all('td')
        if len(cells) > 10:  # Assuming more than 10 columns
            score = cells[10].text.strip()
            Score.append(score)
        if len(cells) > 8:
            coach = cells[8].text.strip()
            Coach.append(coach)

    # Extract Round, Date, and Time for the current season into temporary lists
    for row in rows:
        cells = row.find_all('td')
        if len(cells) > 0:
            match_round = cells[0].text.strip()
            match_date = cells[1].text.strip()
            match_time = cells[2].text.strip()
            temp_round.append(match_round)
            temp_date.append(match_date)
            temp_time.append(match_time)

    # Skip the first 4 items after each iteration
    Match_Round.extend(temp_round[4:])
    Match_Date.extend(temp_date[4:])
    Match_Time.extend(temp_time[4:])

    # Scrape Opponent and Venue (from a different part of the page)
    base_url_opponent = f"https://www.transfermarkt.co.uk/el-ahly-kairo/spielplan/verein/7/saison_id/{season}#EGY1"
    pageTree_opponent = requests.get(base_url_opponent, headers=headers)
    soup_opponent = BeautifulSoup(pageTree_opponent.content, 'html.parser')

    #Away Team
    for row in rows:
        cells = row.find_all('td')
        if len(cells) > 6:
            away = cells[6].text.strip()
            temp_away.append(away)
    Away.extend(temp_away[3:])

    #Home Team
    for row in rows:
        cells = row.find_all('td')
        if len(cells) > 4:
            home = cells[4].text.strip()
            temp_home.append(home)
    Home.extend(temp_home[3:])

    # Extract Formation
    rowss = soup.find_all('tr')
    for row in rowss:
        cells = row.find_all('td')
        if len(cells) > 7:
            formation = cells[7].text.strip()
            temp_formation.append(formation)
    Formation.extend(temp_formation[3:])
    #print season number in list
    Season.extend([season] * len(temp_round[4:]))

    season += 1
    print(f'Season {season} Scrapped')
    if season > 2023:
        break
```

```python
import csv
from itertools import zip_longest
file_list = [Coach, Score, Home, Away, Match_Date , Match_Time, Match_Round, Formation, Season]
exported = zip_longest(*file_list)
with open ("Ahly Matches 2000 to 2024.csv", "w") as myfile:
  wr = csv.writer (myfile)
  wr.writerow(["Coach", "Score", "Home","Away", "Match_Date" , "Match_Time", "Match_Round", "Formation", "Season"])
  wr.writerows(exported)
```
