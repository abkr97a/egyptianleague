```python
from bs4 import BeautifulSoup
import time
import requests
import pandas as pd
import numpy as np
import re
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36'
}
Team_ID = []
Team_Name = []
season = 2000
while True:
  base_url = f"https://www.transfermarkt.com/egyptian-premier-league/startseite/wettbewerb/EGY1/plus/?saison_id={season}"
  pageTree = requests.get(base_url, headers=headers)
  soup = BeautifulSoup(pageTree.content, 'html.parser')
  teams = soup.find_all('td',{'class':'hauptlink no-border-links'})
  for team in teams:
    if team.find('a'):
      href = team.find('a')['href']
      team_id = href.split('/verein/')[1].split('/')[0]
      Team_ID.append(team_id)
      Team_Name.append(team.text.strip())
  season += 1
  print(f'Season {season} Scraped')
  # Stop after scraping up to season 2001
  if season > 2023:
        break
Teams_Data = dict(zip(Team_ID, Team_Name))
print(Teams_Data)
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36'
}
Player_Name = []
Position = []
Age = []
Number = []
InSquad = []
Matches = []
Min = []
Goals = []
Assists = []
Yellow_Cards = []
Red_Cards = []
SubIn = []
SubOut = []
Nationality = []
Season = []
Club_Name = []
Club_ID = []
Player_ID = []

season = 2008

while True:
    for team_id, team_name in Teams_Data.items():
        base_url = f"https://www.transfermarkt.co.uk/zamalek-sc/leistungsdaten/verein/{team_id}/reldata/EGY1%26{season}/plus/1"
        pageTree = requests.get(base_url, headers=headers)
        soup = BeautifulSoup(pageTree.content, 'html.parser')

        # Find all rows
        rows = soup.find_all('tr')

        # Extract Player Data
        names = soup.find_all('td', {'class': 'hauptlink'})
        nationalities = soup.find_all('img', {'class': 'flaggenrahmen'})

        # Player Name Extraction
        for name in names:
            div_tag = name.find('div', {'class': 'di nowrap'})
            if div_tag:
                player_name = div_tag.text.strip()
                Player_Name.append(player_name)
                # Find the <a> tag within the div_tag
                a_tag = div_tag.find('a')
                if a_tag:
                    name_id = a_tag['href']
                    name_id = name_id.split('/profil/spieler/')[1].split('/')[0]
                    Player_ID.append(name_id)


        # Row Data Extraction
        for row in rows:
            cells = row.find_all('td')
            if len(cells) > 10:
                age = cells[5].text.strip()
                Age.append(age)

                position = cells[4].text.strip()
                Position.append(position)

                number = cells[0].text.strip()
                Number.append(number)

                in_squad = cells[7].text.strip()
                InSquad.append(in_squad)

                matches = cells[8].text.strip()
                Matches.append(matches)

                goals = cells[9].text.strip()
                Goals.append(goals)

                assists = cells[10].text.strip()
                Assists.append(assists)

                yellow_cards = cells[11].text.strip()
                Yellow_Cards.append(yellow_cards)

                # Handle Red Cards
                red_card1 = int(cells[12].text.strip()) if cells[12].text.strip().isdigit() else 0
                red_card2 = int(cells[13].text.strip()) if cells[13].text.strip().isdigit() else 0
                red_cards = red_card1 + red_card2
                Red_Cards.append(red_cards)

                sub_in = cells[14].text.strip()
                SubIn.append(sub_in)

                sub_out = cells[15].text.strip()
                SubOut.append(sub_out)

                min_played = cells[17].text.strip()
                Min.append(min_played)

                # Extract Nationality
                img_tag = cells[6].find('img')
                if img_tag and 'title' in img_tag.attrs:
                    nationality = img_tag['title']
                    Nationality.append(nationality)
                else:
                    Nationality.append('non')
                # Append Season and Team Info
                Season.append(season)
                Club_Name.append(team_name)
                Club_ID.append(team_id)

        print(f'Season {season} for {team_name} Scraped')

    # Increment season
    season += 1

    # Stop after scraping up to season 2001
    if season > 2024:
        break
```

```python
import csv
from itertools import zip_longest
file_list = [Season,Club_Name,Club_ID, Player_Name,Player_ID ,Number, Position, Age, InSquad , Matches, Min, Goals, Assists, Yellow_Cards, Red_Cards, Nationality]
exported = zip_longest(*file_list)
with open ("Egyptian Players 2000 to 2024ext.csv", "w") as myfile:
  wr = csv.writer (myfile)
  wr.writerow(["Season","Team Name","Club ID","Player_Name","Player ID", "Number", "Position", "Age", "InSquad" , "Matches", "Min", "Goals", "Assists", "Yellow_Cards", "Red_Cards", "Nationality"])
  wr.writerows(exported)
```
