```python
from ast import Match
headers = {
    'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36'
}

Date = []
Time = []
Home_Team = []
Home_Team_ID = []
Away_Team = []
Away_Team_ID = []
Score = []
Time = []
Season = []
Match_ID = []
Team_ID = []
Goal_Scorers = []
Goal_Scorers_ID = []
Goal_ID = []
Match_URLS = []
season = 2000
while True:
    base_url = f"https://www.transfermarkt.co.uk/egyptian-premier-league/gesamtspielplan/wettbewerb/EGY1?saison_id={season}&spieltagVon=1&spieltagBis=34"
    pageTree = requests.get(base_url, headers=headers)
    soup = BeautifulSoup(pageTree.content, 'html.parser')



    datestime = soup.find_all('tr')
    for date in datestime:
      row = date.find_all('td',{'class':'hide-for-small'}) # Call find_all on the date variable
      if row:
        date = row[0].text.strip()
        time = row[1].text.strip()
        Date.append(date)
        Time.append(time)

    hometeams = soup.find_all('td',{'class':'text-right no-border-rechts hauptlink'})
    awayteams = soup.find_all('td',{'class':'no-border-links hauptlink'})
    scores = soup.find_all('td',{'class':'zentriert hauptlink'})

    for hometeam in hometeams:
      if hometeam.find('a'):
        hrefh = hometeam.find('a')['href']
        teamh_id = hrefh.split('/verein/')[1].split('/')[0]
        teamh = hometeam.find('a')['title']
        Home_Team_ID.append(teamh_id)
        Home_Team.append(teamh)

    for awayteam in awayteams:
      if awayteam.find('a'):
        hrefa = awayteam.find('a')['href']
        teama_id = hrefa.split('/verein/')[1].split('/')[0]
        teama = awayteam.find('a')['title']
        Away_Team_ID.append(teama_id)
        Away_Team.append(teama)

    for score in scores:
      Score.append(score.text.strip())
      Season.append(season)
      if score.find('a'): # Check if an 'a' tag exists before accessing it
        url = score.find('a').get('href')
        id = url.split('/')[4]
        Match_ID.append(id)

      else:
        score = None # Or handle the case where there is no 'a' tag
      Match_URLS.append(url)
    print (season, "Scraped")

    season += 1
    if season > 2023:
        break
```

```python
df = pd.read_csv('/content/Matches 2000 to 2024.csv')
df.fillna(method='ffill', inplace=True)
df.head(50)
```
